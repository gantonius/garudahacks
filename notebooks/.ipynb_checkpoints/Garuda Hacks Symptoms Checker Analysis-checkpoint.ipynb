```python
covid_sample = pd.read_csv('hack\explatory\covid_sampled_data.csv')
covid_sample.drop(columns = ['Unnamed: 0'], inplace = True)
covid_sample.set_index('patient', inplace = True)
covid_sample
```

```python
X = covid_sample.drop(columns = ['sick'])
y = covid_sample['sick']
```

```python
sse = []
k_range = range(1, 15)
for i in k_range:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(covid_sample)
    sse.append(kmeans.inertia_)
```

```python
sse
```

```python
%matplotlib inline
sns.set()

plt.figure(figsize = (8, 8))
plt.plot(k_range, sse)
plt.xlabel('K values')
plt.ylabel('Standard Summation Error')
plt.show() # Plotting the elbow plot
```

```python
# Path needs to be changed to wherever the KMeans model is in
def severity_classifier(symptoms, PATH = 'k-means_model.joblib'):
    """
    Function that predicts whether the patient has a high severity COVID case or a low severity
    COVID case if the patient is confirmed to be tested positive for COVID-19
    """
    import joblib
    
    symptoms = list(symptoms)
    symptoms.append(1) # Appending the sick binary value
    
    classifier = joblib.load(PATH)
    severity_id = classifier.predict([symptoms])[0] + 1
    
    if severity_id == 1:
        severity = 'High Severity'
    else:
        severity = 'Low Severity'
    
    return severity
```

```python
symptoms_df = pd.read_csv('disease_clean.csv')
symptoms_df.drop(columns = 'Unnamed: 0', inplace = True)
symptoms_df.set_index('patient', inplace = True)
symptoms_df['Sick'] = symptoms_df['Sick'].apply(lambda x: np.where(x == 1, 'Covid-19', np.where(x == 2, 'Dengue', 'Tuberculosis')))
symptoms_df
```

```python
X = symptoms_df.drop(columns = ['Sick'])
y = symptoms_df['Sick']
```

```python
log_model = LogisticRegression()
log_model.fit(X, y)
```

```python
joblib.dump(log_model, 'logistic-reg_model.joblib')
```

```python
# Path needs to be changed to whatever directory the Logistic Regression model is in
def disease_predictor(symptoms, PATH = "logistic-reg_model.joblib"):
    """
    Function that predicts whether a patient has COVID-19, Tuberculosis, or Dengue Fever based
    on a list of 18 symptoms
    """
    import joblib
    log_model = joblib.load(PATH)
    
    disease = log_model.predict([symptoms])[0]
    
    return disease
```
